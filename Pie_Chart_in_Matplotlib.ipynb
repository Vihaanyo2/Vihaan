```python
import matplotlib.pyplot as plt

categories = ['Apple', 'Banana', 'Cherry', 'Grapes']
values = [30, 20, 40, 10]
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90)

plt.title('Fruit Distribution')
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
categories = ["Category A", "Category B", "Category C", "Category D"]
sizes = [25, 40, 15, 20]
max_size = max(sizes)
max_index = sizes.index(max_size)

explode_value = 0.1 # This value determines how far the slice is pulled out
explode = [0] * len(categories)
explode[max_index] = explode_value
plt.figure(figsize=(8, 8)) # Optional: set the figure size for better display
plt.pie(sizes, labels=categories, explode=explode, autopct='%1.1f%%', startangle=90, shadow=True)

plt.title("Pie Chart with Largest Slice Exploded")

plt.axis('equal')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Using data and labels from a previous example
data_for_slices = [15, 30, 45, 10]
slice_labels = ['Category A', 'Category B', 'Category C', 'Category D']

plt.pie(
    data_for_slices,  # Your data for pie slices
    labels=slice_labels,
    startangle=90,     # Rotates the chart to start at the top
    autopct='%1.1f%%' # Added autopct for displaying percentages
)
plt.title('Fixed Pie Chart Example') # Added a title
plt.axis('equal') # Added to ensure the pie chart is circular
plt.show()
```

```python
import matplotlib.pyplot as plt

sizes = [45, 25, 15, 15]
labels = ['Team A', 'Team B', 'Team C', 'Team D']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

my_circle = plt.Circle((0, 0), 0.7, color='white')
p = plt.gcf()
p.gca().add_artist(my_circle)

plt.axis('equal')
plt.title('Donut Chart Example')
plt.show()
```

```python
import matplotlib.pyplot as plt

sizes = [30, 25, 20, 15, 10]
labels = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgrey']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.legend(loc="upper right") # You can adjust the location as needed
plt.axis('equal')

plt.show()
```

```python
import matplotlib.pyplot as plt

sizes = [15, 30, 45, 10]  # Proportions of each slice
labels = ['Category A', 'Category B', 'Category C', 'Category D'] # Labels for each slice
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'] # Colors for each slice
explode = (0, 0.1, 0, 0)  # "Explode" the second slice (Category B) for emphasis

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.savefig("pie_chart.png")

plt.show()
```
