```python
import numpy as np
```

```python
array_1_to_20 = np.arange(1, 21)
print(f"The array is: {array_1_to_20}\n")
```

```python
array_sum = np.sum(array_1_to_20)
array_mean = np.mean(array_1_to_20)
array_median = np.median(array_1_to_20)
array_std = np.std(array_1_to_20)

print(f"Sum of elements: {array_sum}")
print(f"Mean of elements: {array_mean}")
print(f"Median of elements: {array_median}")
print(f"Standard deviation of elements: {array_std}\n")
```

```python
mask_greater_than_10 = array_1_to_20 > 10
```

```python
indices_greater_than_10 = np.where(mask_greater_than_10)
```

```python
indices_list = indices_greater_than_10[0]

print(f"Indices of elements greater than 10: {indices_list}")
print(f"Elements at these indices: {array_1_to_20[indices_list]}")
```

```python
array_2d = array_1_to_20.reshape(4, 5)
transpose_array = array_2d.T
print("Original Array:\n", array_2d)
print("Transpose of the Array:\n", transpose_array)
```

```python
row_sums = array_2d.sum(axis=1)
col_sums = array_2d.sum(axis=0)
print("Row-wise Sums:", row_sums)
print("Column-wise Sums:", col_sums)
```

```python
array1 = np.random.randint(1, 21, size=(3, 3))
array2 = np.random.randint(1, 21, size=(3, 3))
addition_result = array1 + array2
subtraction_result = array1 - array2
multiplication_result = array1 * array2
dot_product_result = np.dot(array1, array2)

print(f"Array 1:\n{array1}")
print(f"Array 2:\n{array2}")
print(f"Element-wise Addition:\n{addition_result}")
print(f"Element-wise Subtraction:\n{subtraction_result}")
print(f"Element-wise Multiplication:\n{multiplication_result}")
print(f"Dot Product:\n{dot_product_result}")
```

```python
arr_1d = np.arange(1, 13)
print("Original 1D array:")
print(arr_1d)
arr_2d = arr_1d.reshape(3, 4)
print("\nReshaped 3x4 2D array:")
print(arr_2d)
sliced_arr = arr_2d[:2, 2:]
print("\nSliced array (first two rows, last two columns):")
print(sliced_arr)
```
