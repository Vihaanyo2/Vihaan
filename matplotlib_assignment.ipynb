```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', linestyle='-')
plt.title('Simple Line Plot of X vs Y')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True)
plt.show()
```

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]
colors = ['skyblue', 'salmon', 'lightgreen', 'gold']
plt.bar(students, marks, color=colors)
plt.xlabel('Students')
plt.ylabel('Marks Scored')
plt.title('Student Marks in Subject')
plt.show()
```

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
max_revenue = max(revenue)
max_index = revenue.index(max_revenue)
explode = [0] * len(regions)
explode[max_index] = 0.1
plt.figure(figsize=(8, 6)) # Optional: adjust figure size
plt.pie(revenue, labels=regions, explode=explode, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Company Revenue Distribution by Region')
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
sample_size = 1000
min_val = 1
max_val = 100
random_integers = np.random.randint(min_val, max_val + 1, sample_size)
plt.hist(random_integers, bins=range(min_val, max_val + 2), color='skyblue', edgecolor='black')
plt.xlabel('Integer Value')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of 1000 Random Integers (1-100)')
plt.grid(True)
plt.show()
```
