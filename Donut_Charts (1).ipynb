```python
import matplotlib.pyplot as plt
departments = ['Sales', 'Marketing', 'Engineering', 'HR']
employees = [150, 100, 200, 75]
plt.pie(employees, labels=departments, autopct='%1.1f%%', startangle=90) # autopct shows percentages
my_circle = plt.Circle((0,0), 0.7, color='white')
p = plt.gcf()
p.gca().add_artist(my_circle)
plt.title("Employee Distribution by Department")
plt.show()
```

```python
import matplotlib.pyplot as plt
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 35, 20, 20]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Categories')
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
sizes = [15, 30, 45, 10]
labels = ['Apples', 'Oranges', 'Bananas', 'Grapes']
plt.pie(sizes, labels=labels, wedgeprops=dict(width=0.4))
plt.axis("equal")
plt.title("Fruit Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt
sizes = [15, 30, 45, 10]
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(aspect="equal"))
wedges, texts = ax.pie(sizes, colors=colors, wedgeprops=dict(width=0.4), startangle=90)
ax.legend(wedges, labels, title="Animals", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
ax.set_title("Donut Chart Thickness Example")
plt.show()
```

```python
import matplotlib.pyplot as plt
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [20, 35, 15, 30]
custom_colors = ['#4c78a8', 'skyblue', '#6b92bc', 'darkorange']
fig, ax = plt.subplots(figsize=(8, 6))
wedges, texts = ax.pie(
    sizes,
    wedgeprops=dict(width=0.5),  # Create the donut hole effect
    startangle=90,
    colors=custom_colors
)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)
ax.set_title("Custom Colored Donut Chart")
ax.legend(wedges, labels, title="Categories", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from 'Book1.csv'
df = pd.read_csv('Book1.csv')

# Sort the DataFrame by 'Population' in descending order
df_sorted = df.sort_values('Population', ascending=False)

# Create a bar chart
plt.bar(df_sorted['City'], df_sorted['Population'])

# Add labels and title
plt.xlabel('City')
plt.ylabel('Population')
plt.title('Cities by Population')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Department A', 'Department B', 'Department C', 'Department D']
sizes = [25, 35, 20, 20]
max_employees = max(sizes)
max_index = sizes.index(max_employees)
explode = tuple([0.0 if i != max_index else 0.1 for i in range(len(sizes))])

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Employee Distribution by Department')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
labels = ['Category A', 'Category B', 'Category C', 'Category D']
values = [30, 25, 20, 25]
fig, ax = plt.subplots()
ax.pie(values, labels=labels, autopct='%1.1f%%', pctdistance=0.85,
       wedgeprops=dict(width=0.4, edgecolor='white'))
centre_circle = plt.Circle((0,0),0.60,fc='white')
fig.gca().add_artist(centre_circle)
ax.axis('equal')
plt.savefig("donut chart.png")
plt.show()
```
