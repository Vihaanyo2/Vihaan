```python
employees = {
    'John': {'age': 54, 'salary': 67000},
    'Jane': {'age': 42, 'salary': 43000},
    'Bob': {'age': 49, 'salary': 55000},
    'Alice': {'age': 57, 'salary': 71000},
    'Mike': {'age': 35, 'salary': 25000}
}

total_age = sum(emp['age'] for emp in employees.values())
total_salary = sum(emp['salary'] for emp in employees.values())
num_employees = len(employees)

average_age = total_age / num_employees
average_salary = total_salary / num_employees

oldest_employee = max(employees, key=lambda name: employees[name]['age'])
youngest_employee = min(employees, key=lambda name: employees[name]['age'])

print(f"Average Age: {average_age:.2f}")
print(f"Average Salary: ${average_salary:,.2f}")
print(f"Oldest Employee: {oldest_employee} (Age: {employees[oldest_employee]['age']})")
print(f"Youngest Employee: {youngest_employee} (Age: {employees[youngest_employee]['age']})")
```

```python
from datetime import datetime

employees = [
    {
        "name": "Alice",
        "employee_id": 101,
        "joining_date": datetime(2013, 5, 15)
    },
    {
        "name": "Bob",
        "employee_id": 102,
        "joining_date": datetime(2016, 8, 20)
    },
    {
        "name": "Charlie",
        "employee_id": 103,
        "joining_date": datetime(2012, 1, 10)
    },
    {
        "name": "David",
        "employee_id": 104,
        "joining_date": datetime(2018, 11, 5)
    },
    {
        "name": "Eve",
        "employee_id": 105,
        "joining_date": datetime(2014, 7, 22)
    },
]

def get_employees_before_2015(employee_list):
    old_employees = []
    for employee in employee_list:
        if employee["joining_date"].year < 2015:
            old_employees.append(employee)
    return old_employees

old_employees = get_employees_before_2015(employees)
print(old_employees)
```

```python
employees = [
    {"name": "Alice", "salary": 75000},
    {"name": "Bob", "salary": 90000},
    {"name": "Charlie", "salary": 60000},
    {"name": "David", "salary": 110000},
    {"name": "Eve", "salary": 80000},
    {"name": "Frank", "salary": 95000},
    {"name": "Grace", "salary": 120000},
    {"name": "Heidi", "salary": 70000},
    {"name": "Ivan", "salary": 105000},
    {"name": "Judy", "salary": 85000},
]

sorted_employees = sorted(employees, key=lambda employee: employee["salary"], reverse=True)

top_5_employees = sorted_employees[:5]

print("Top 5 Highest-Paid Employees:")
for employee in top_5_employees:
    print(f"Name: {employee['name']}, Salary: ${employee['salary']:,}")
```
