```python

import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 20, 13]

plt.plot(x, y, '-o')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Graph')

plt.show()
```

```python
import matplotlib.pyplot as plt
x1 = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]

x2 = [1, 2, 3, 4, 5]
y2 = [1, 4, 9, 16, 25]

plt.plot(x1, y1, marker='^', linestyle='-', label='Line 1 (^)')

plt.plot(x2, y2, marker='s', linestyle='-', label='Line 2 (s)')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Two Lines with Different Markers')
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(8, 6)) # Optional: set figure size
plt.scatter(x, y, marker='*', s=100, color='blue', label='Random Points')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot of 50 Random Points with Star Markers")

plt.legend()

plt.grid(True) # Optional: add a grid
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 5, 8, 3, 10]

plt.plot(x, y,
         marker='D',         # 'D' for a diamond marker
         markersize=12       # Size of the marker
        )

plt.title('Line Graph with Diamond Markers')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3, 4])
y1 = np.array([0, 1, 4, 9, 16])

x2 = np.array([0, 1, 2, 3, 4])
y2 = np.array([16, 9, 4, 1, 0])

x3 = np.array([0, 1, 2, 3, 4])
y3 = np.array([8, 4, 2, 1, 0.5])

plt.figure()

# Plot the first line with circle markers
plt.plot(x1, y1, marker='o', linestyle='-', label='Circle Marker')

plt.plot(x2, y2, marker='p', linestyle='--', label='Pentagon Marker')

plt.plot(x3, y3, marker='x', linestyle=':', label='Cross Marker')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot with Different Marker Styles")
plt.legend()

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 5, 2, 7, 9]
y3 = [8, 6, 7, 3, 5]

plt.plot(x, y1, label='Line 1 (Squares)', marker='s', linestyle='-', color='blue')
plt.plot(x, y2, label='Line 2 (Circles)', marker='o', linestyle='--', color='green')
plt.plot(x, y3, label='Line 3 (Triangles)', marker='^', linestyle=':', color='red')

plt.title('Multi-Line Graph with Unique Markers')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend()

plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
revenue = [50000, 55000, 60000, 58000, 62000, 65000, 70000, 68000, 72000, 75000, 80000, 85000]

plt.figure(figsize=(10, 6)) # Optional: Adjust figure size for better readability
plt.plot(months, revenue, marker='o') # 'o' adds circular markers for data points
plt.title("Sales Data Visualization")
plt.xlabel("Month")
plt.ylabel("Revenue ($)")
plt.grid(True)

plt.show()
```
