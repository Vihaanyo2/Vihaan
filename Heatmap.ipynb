```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = {
    'Math': [85, 78, 92, 65, 70],
    'Science': [90, 82, 88, 75, 68],
    'English': [70, 85, 75, 80, 95],
    'History': [60, 72, 80, 88, 77]
}
student_names = ['Alice', 'Bob', 'Charlie', 'David', 'Eve']
df_marks = pd.DataFrame(data, index=student_names)

print("Student Marks Data:")
print(df_marks)

plt.figure(figsize=(8, 6)) # Adjust figure size for better readability
sns.heatmap(df_marks, annot=True, cmap='viridis', fmt='g')

plt.title('Student Marks Heatmap')
plt.xlabel('Subjects')
plt.ylabel('Students')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = np.random.rand(10, 10)
df = pd.DataFrame(data, columns=[f'Col_{i}' for i in range(10)], index=[f'Row_{i}' for i in range(10)])

plt.figure(figsize=(8, 6))
sns.heatmap(df, cmap='coolwarm', annot=True, fmt=".2f", linewidths=.5)
plt.title('Heatmap with "coolwarm" Colormap')
plt.show()

plt.figure(figsize=(8, 6))
sns.heatmap(df, cmap='magma', annot=True, fmt=".2f", linewidths=.5)
plt.title('Heatmap with "magma" Colormap')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = {'Student': ['Alice', 'Bob', 'Charlie', 'David'],
        'Math': [85, 70, 92, 65],
        'Science': [90, 75, 88, 70],
        'English': [78, 82, 95, 60]}
df = pd.DataFrame(data)


df['Average_Marks'] = df.drop('Student', axis=1).mean(axis=1)
df_sorted = df.sort_values(by='Average_Marks', ascending=False)
heatmap_data = df_sorted.set_index('Student').drop('Average_Marks', axis=1)


plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, annot=True, cmap='viridis', fmt=".0f", linewidths=.5)
plt.title('Student Marks Heatmap (Sorted by Average Marks)')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Student_ID': ['S1', 'S2', 'S3', 'S4', 'S5'],
    'Math': [88, 92, 75, 80, 95],
    'Physics': [90, 85, 78, 88, 91],
    'Chemistry': [78, 88, 85, 90, 84],
    'Biology': [95, 79, 88, 85, 93]
}
df = pd.DataFrame(data).set_index('Student_ID')

print("Original DataFrame:")
print(df)
print("\n" + "="*30 + "\n")

df_transposed = df.T

print("Transposed DataFrame:")
print(df_transposed)
print("\n" + "="*30 + "\n")

plt.figure(figsize=(10, 6))
sns.heatmap(df_transposed, annot=True, cmap='coolwarm', fmt="d", linewidths=.5)

plt.title('Heatmap of Student Scores by Subject')
plt.xlabel('Student')
plt.ylabel('Subject')

plt.show()
```
