```python
import matplotlib.pyplot as plt

departments = ['HR', 'Engineering', 'Marketing', 'Sales', 'Finance']
employees = [30, 120, 50, 80, 40]

fig, ax = plt.subplots()

wedges, texts, autotexts = ax.pie(employees, labels=departments, autopct='%1.1f%%',
                                  startangle=90, pctdistance=0.85,
                                  wedgeprops={'width': 0.4})

ax.axis('equal')

plt.title('Number of Employees per Department')

plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 35, 20, 20] # Example values

plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90)

plt.title("Distribution of Categories")

plt.axis('equal')

plt.show()
```

```python
import matplotlib.pyplot as plt

department_sizes = [30, 25, 15, 20, 10]
department_labels = ['Sales', 'Engineering', 'Marketing', 'Operations', 'HR']

fig, ax = plt.subplots()
ax.axis('equal')

wedges, texts, autotexts = ax.pie(
    department_sizes,
    labels=department_labels,
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops={'width': 0.3}
)

ax.set_title('Employee Distribution by Department')

ax.legend(wedges, department_labels, title="Departments", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

sizes = [30, 25, 20, 15, 10]
labels = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'pink']

fig, ax = plt.subplots()

wedges, texts, autotexts = ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%',
                       startangle=90, pctdistance=0.85)

centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)

plt.axis("equal") # This line is crucial for a perfectly round chart

ax.set_title("Perfectly Round Donut Chart")

plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']
sizes = [25, 30, 20, 25]

fig, ax = plt.subplots(subplot_kw=dict(aspect="equal"))

wedges, texts = ax.pie(sizes,
                       labels=labels,
                       wedgeprops={"width": 0.4})

ax.set_title("Donut Chart with Customized Thickness")

plt.show()
```

```python
import plotly.express as px
import pandas as pd

data = {'labels': ['Category A', 'Category B', 'Category C', 'Category D'],
        'values': [30, 20, 15, 35]}
df = pd.DataFrame(data)

custom_colors = ['#EF553B', '#636EFA', '#00CC96', '#AB63FA']

fig = px.pie(
    df,
    values='values',
    names='labels',
    color_discrete_sequence=custom_colors,
    hole=0.5 # Set `hole` to a value between 0 and 1 to create a donut chart
)

fig.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

departments = ['Sales', 'Engineering', 'Marketing', 'Finance']
employees = [150, 400, 100, 120]

max_employees = max(employees)
max_index = employees.index(max_employees)

explode = [0] * len(departments)
explode[max_index] = 0.1

fig, ax = plt.subplots(figsize=(8, 8))

ax.pie(employees,
       explode=explode,
       labels=departments,
       autopct='%1.1f%%',  # Display percentages
       shadow=True,
       startangle=90)
ax.axis('equal')  # Ensure the pie chart is a perfect circle

plt.title('Employee Distribution by Department')
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 30, 20, 25]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

fig, ax = plt.subplots()

ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%',
       startangle=90, pctdistance=0.85,
       wedgeprops=dict(width=0.4, edgecolor='w'))

centre_circle = plt.Circle((0,0),0.60,fc='white')
fig.gca().add_artist(centre_circle)

ax.axis('equal')

plt.savefig("donut chart.png")

plt.show()
```
