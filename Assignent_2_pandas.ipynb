```python
import pandas as pd
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}
```

```python
df = pd.DataFrame(data)
```

```python
print("First five rows of the DataFrame:")
print(df.head())
print("\n")
```

```python
print("Summary statistics for Age and Salary:")
print(df[['Age', 'Salary']].describe())
print("\n")
```

```python
hr_salaries = df[df['Department'] == 'HR']['Salary']
average_hr_salary = hr_salaries.mean()
print(f"Average salary of employees in the HR department: ${average_hr_salary:,.2f}")
```

```python
df['Bonus'] = df['Salary'] * 0.1
```

```python
print(df)
```

```python
filtered_df = df[(df['Age'] >= 25) & (df['Age'] <= 30)]

print(filtered_df)
```

```python
avg_salary = df.groupby('Department')['Salary'].mean()

print(avg_salary)
```

```python
sorted_df = df.sort_values(by='Salary', ascending=True)
sorted_df.to_csv('sorted_employees.csv', index=False)

print(sorted_df)
```
