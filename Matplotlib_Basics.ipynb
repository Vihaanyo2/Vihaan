```python
import matplotlib.pyplot as plt

categories = ['Electronics', 'Clothing', 'Groceries']
sales = [15000, 25000, 18000]
colors = ['skyblue', 'lightcoral', 'lightgreen']

plt.figure(figsize=(8, 6)) # Set the figure size for better readability
plt.bar(categories, sales, color=colors)

plt.title('Sales Data by Category')
plt.xlabel('Categories')
plt.ylabel('Sales ($)')

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

num_points = 50
x_coords = np.random.rand(num_points)
y_coords = np.random.rand(num_points)

plt.scatter(x_coords, y_coords, color='red')

# Add a title to the plot
plt.title('Random Scatter Plot')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(size=1000)

plt.hist(data, bins=30, color='purple', edgecolor='black')

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Normally Distributed Data")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [100, 120, 150, 130, 160, 180, 200, 220, 250, 230, 280, 300]

plt.figure(figsize=(10, 6)) # Optional: set figure size
plt.plot(months, sales, marker='o') # Add marker for data points

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C', 'D']
values = [10, 25, 15, 30]
x_data = np.linspace(0, 10, 100)
y_data = np.sin(x_data)

fig, axs = plt.subplots(1, 2, figsize=(10, 4)) # Adjust figsize as needed
axs[0].bar(categories, values, color='skyblue')
axs[0].set_title('Bar Chart Example')
axs[0].set_xlabel('Category')
axs[0].set_ylabel('Value')

axs[1].plot(x_data, y_data, color='salmon', linestyle='--')
axs[1].set_title('Line Plot Example')
axs[1].set_xlabel('X-axis')
axs[1].set_ylabel('Y-axis')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Product A', 'Product B', 'Product C']
sizes = [40, 35, 25]
colors = ['#ff9999','#66b3ff','#99ff99']

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Percentage Distribution of Products")
plt.legend(labels, loc="best")
plt.show()
```
