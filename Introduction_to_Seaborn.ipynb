```python
import pandas as pd
import matplotlib.pyplot as plt
data = {'Age': [25, 30, 35, 40, 45, 50, 55, 60],
            'Salary': [50000, 60000, 75000, 90000, 110000, 125000, 130000, 135000]}
df = pd.DataFrame(data)
plt.figure(figsize=(10, 6)) # Optional: Adjust figure size
plt.plot(df['Age'], df['Salary'], marker='o', linestyle='-')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Salary Trend by Age')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
age = np.random.randint(22, 65, 100)
salary = 30000 + age * 1500 + np.random.normal(0, 10000, 100)
salary[age > 45] += (age[age > 45] - 45) * 500
data = pd.DataFrame({'Age': age, 'Salary': salary})
plt.figure(figsize=(10, 6))
plt.scatter(data['Age'], data['Salary'], alpha=0.7)
plt.title('Age vs Salary Scatterplot')
plt.xlabel('Age (Years)')
plt.ylabel('Salary (USD)')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = {'Department': ['Sales', 'Technical', 'HR', 'Sales', 'Support', 'Technical', 'HR', 'Sales', 'Technical']}
df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
sns.countplot(x='Department', data=df)
plt.title('Number of Employees in Each Department')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {'Category': ['A', 'B', 'C', 'D', 'E'],
        'Value': [15, 25, 30, 20, 10]}
df = pd.DataFrame(data)
sns.set_palette("Set2")
sns.barplot(x='Category', y='Value', data=df)
plt.title('Bar Chart with "Set2" Color Palette')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = {'Department': ['HR', 'IT', 'Sales', 'HR', 'IT', 'Sales', 'HR', 'IT', 'Sales'],
        'Salary': [50000, 70000, 60000, 55000, 75000, 62000, 52000, 72000, 65000]}
df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Department', y='Salary', data=df)
plt.title('Salary Distribution Across Departments')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure(figsize=(8, 6)) # Optional: set figure size
plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.savefig("plot.png", dpi=300, bbox_inches='tight')
plt.show()
```
