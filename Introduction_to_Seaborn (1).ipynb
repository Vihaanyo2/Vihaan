```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Department': ['HR', 'Engineering', 'Marketing', 'HR', 'Engineering', 'Marketing', 'HR', 'Engineering'],
    'Salary': [60000, 90000, 70000, 65000, 95000, 72000, 62000, 88000]
}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6)) # Optional: Adjust figure size
sns.barplot(x='Department', y='Salary', data=df, estimator='mean')

plt.title("Average Salary by Department")

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Age': [22, 25, 28, 30, 35, 40, 45, 50, 55, 60],
    'Salary': [30000, 35000, 42000, 50000, 65000, 80000, 95000, 110000, 120000, 125000]
}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6)) # Optional: Adjust figure size
plt.plot(df['Age'], df['Salary'], marker='o', linestyle='-')

plt.title('Salary Trend by Age')
plt.xlabel('Age')
plt.ylabel('Salary')

plt.grid(True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = {
        'Age': [25, 30, 35, 40, 45, 50, 28, 33, 38, 42, 48, 55],
        'Salary': [50000, 60000, 75000, 90000, 110000, 130000, 55000, 68000, 85000, 100000, 120000, 140000],
        'Department': ['HR', 'IT', 'Finance', 'IT', 'HR', 'Finance', 'IT', 'HR', 'Finance', 'IT', 'HR', 'Finance']
}
df = pd.DataFrame(data)
plt.figure(figsize=(10, 6)) # Optional: set figure size
sns.scatterplot(data=df, x='Age', y='Salary', hue='Department', s=100) # s for point size
plt.title('Age vs Salary by Department')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True) # Optional: add a grid
plt.legend(title='Department') # Add a legend for the departments
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Department': ['HR', 'IT', 'Sales', 'HR', 'IT', 'Marketing', 'Sales', 'IT', 'HR', 'Marketing'],
    'Employee_ID': range(1, 11)
}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6)) # Adjust figure size for better readability
sns.countplot(x='Department', data=df, palette='viridis') # Using 'viridis' palette for visual appeal

plt.title('Number of Employees per Department', fontsize=16)
plt.xlabel('Department', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)

plt.xticks(rotation=45, ha='right')

plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
sns.scatterplot(
    data=iris,
    x="sepal_length",
    y="petal_length",
    hue="species",
    palette="Set2"
)

plt.title("Iris Species by Sepal and Petal Length (Set2 Palette)")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Department': ['HR', 'IT', 'Sales', 'HR', 'IT', 'Sales', 'HR', 'IT', 'Sales', 'HR', 'IT', 'Sales'],
    'Salary': [50000, 75000, 60000, 55000, 80000, 62000, 52000, 78000, 65000, 120000, 30000, 70000]
}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6)) # Optional: Adjust figure size for better readability
sns.boxplot(x='Department', y='Salary', data=df)

plt.title('Salary Distribution by Department')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

print("First 5 rows of the tips dataset:")
print(tips.head())

plt.figure(figsize=(8, 6))
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Scatterplot of Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.grid(True)
plt.show()
```
